```python
import pandas as pd
import python_speech_features as psf
import numpy as np
import os
```

```python
records = pd.read_csv("the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data.csv")
records
```

```python
patient_id = list(records["Patient ID"])
recording_loc = list(records["Recording locations:"])
murmur = list(records["Murmur"])
murmur_loc = list(records["Murmur locations"])
```

```python
recording_loc = [x.split("+") for x in recording_loc]
for i in range(len(murmur)):
    if murmur[i] == 'Absent':
        murmur[i] = 0
    elif murmur[i] == 'Present':
        murmur[i] = 1
    else:
        murmur[i] = 2

for i in range(len(murmur_loc)):
    if murmur_loc[i] is np.nan:
        murmur_loc[i] = []
    else:
        murmur_loc[i] = murmur_loc[i].split("+")
```

```python
import xlwt
import scipy.io.wavfile as wav

for i in range(len(patient_id)):
    for loc in recording_loc[i]:
        if os.path.isfile("the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data/" + str(patient_id[i]) + "_" + loc + ".wav"):
            rate, signal = wav.read("the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data/" + 
                                    str(patient_id[i]) + "_" + loc + ".wav")
            features = psf.base.mfcc(signal, samplerate = rate)
        else:
            rate, signal = wav.read("the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data/" + 
                                    str(patient_id[i]) + "_" + loc + "_1.wav")
            features = psf.base.mfcc(signal, samplerate = rate)
        np.savetxt("SSE/" + str(patient_id[i]) + "_" + loc + "_MFCC.csv", features, delimiter=',')
    print("Completed patient #", str(i+1))
```
