```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Embedding
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data.csv")
df
```

```python
patient_id = list(df["Patient ID"])
recording_loc = list(df["Recording locations:"])
murmur = list(df["Murmur"])
murmur_loc = list(df["Murmur locations"])
df.columns
```

```python
recording_loc = [x.split("+") for x in recording_loc]

for i in range(len(murmur_loc)):
    if murmur_loc[i] is np.nan:
        murmur_loc[i] = []
    else:
        murmur_loc[i] = murmur_loc[i].split("+")
```

```python
X = []
y = []

for i in range(len(patient_id)):
    if murmur[i] == "Absent":
        for recloc in recording_loc[i]:
            X.append(np.loadtxt("SSE/" + str(patient_id[i]) + "_" + recloc + "_features.csv", delimiter=',')[:200])
            y.append(0)
    elif murmur[i] == "Present":
        for recloc in recording_loc[i]:
            if recloc in murmur_loc[i]:
                X.append(np.loadtxt("SSE/" + str(patient_id[i]) + "_" + recloc + "_features.csv", delimiter=',')[:200])
                y.append(1)
            else:
                X.append(np.loadtxt("SSE/" + str(patient_id[i]) + "_" + recloc + "_features.csv", delimiter=',')[:200])
                y.append(0)
    print("Completed patient #", str(i+1))
```

```python
time_steps = 200
feature_length = 18
batch_size = 3
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.33)
```

```python
print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
def reshape_flatten(X):
    output = []
    for row in X:
        for i in range(18):
            output.append(row[i])
    return np.array(output)


X_train_flattened = [reshape_flatten(sample) for sample in X_train]
print(X_train_flattened[0])
print(X_train[0])
X_train_flattened = np.array(X_train_flattened)
y_train = np.array(y_train)
print("------------------")
print(X_train_flattened)
print(y_train)
print(X_train_flattened.shape)
print(y_train.shape)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train_flattened, y_train)
```

```python
X_res.shape
```

```python
y_res.shape
```

```python
model = Sequential()
model.add(LSTM(32))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
def reshape_unflatten(X):
    output = []
    new_X = list(X)
    for i in range(200):
        output.append([])
        for j in range(18):
            output[i].append(new_X[18*i + j])
    return np.array(output)

X_train_new = np.array([reshape_unflatten(i) for i in X_res])
print(X_train_new)
print("-----------------------------------")
print(X_res)
X_train_new.shape
```

```python
y_res.shape
```

```python
model.fit(X_train_new, y_res, epochs=10, batch_size=batch_size)
```

```python
final_final_predicts = model.predict(X_test)
final_final_predicts.shape
```

```python
X = np.array(X)
y = np.array(y)
score_full = model.evaluate(X, y, batch_size=batch_size, verbose=0)
print('Test loss:', score_full[0])
print('Accuracy:', score_full[1])
```

```python
score = model.evaluate(X_test, y_test, batch_size=batch_size, verbose=0)
print('Test loss:', score[0])
print('Accuracy:', score[1])
```
